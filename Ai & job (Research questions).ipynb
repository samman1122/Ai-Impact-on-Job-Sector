```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
#Load the dataset
df=pd.read_csv("Downloads/ai_job_impact.csv")
df.head()
```

### Q1:How can we support employees whose productivity has decreased to improve their performance?

For this first of all we filter the employes with decreased productivity, then check their characteristics
such as (work hours, Ai adoption level, job satisfaction, work type(remote yes/no) )

```python
#Filter decreased productivty employees
low_pro=df[df['Productivity_Change_%'] < 0]

#print(low_pro)
print('The number of Employees with productivity decreased are:',len(low_pro))
```

```python
#Check/Analyze Characteristics

#JOB SATISFACTION
plt.figure(figsize=(6,4))

ax=sns.countplot(x='Job_Satisfaction' , data=low_pro,palette='bright')

# Add labels
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width()/2, p.get_height()),
                ha='center', va='bottom')


plt.title('Job Satisfaction (Low Productivity)')
plt.show()
```

```python
#WORK HOURS
plt.figure(figsize=(6,4))

sns.histplot(low_pro["Work_Hours_Per_Week"], bins=10, kde=True)

plt.title("Work Hours (Low Productivity Employees)")
plt.show()
```

```python
#WORK TYPE
plt.figure(figsize=(6,4))

ax=sns.countplot(x="Remote_Work", data=low_pro)

#Add Labels
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width()/2, p.get_height()),
                ha='center', va='bottom')

plt.title("Work Type (Low Productivity Employees)")
plt.show()
```

```python
plt.figure(figsize=(6,4))

ax=sns.countplot(x="AI_Adoption_Level", data=low_pro)

#Add labels
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width()/2, p.get_height()),
                ha='center', va='bottom')

plt.title("Adoption Level (Low Productivity Employees)")
plt.show()
```

## Conclusion:

The decrease in productivity is not driven by a single factor such as job satisfaction, work type, or
adoption level, as these are evenly distributed among employees. However, work hours appear slightly
higher, suggesting that workload and fatigue may be contributing factors to reduced productivity.


## Short Line:
Low productivity is not linked to satisfaction,ai_adoption or work type, but higher working hours may be affecting employee performance.


## Business Insight
Since productivity decline is spread across all groups, the company should focus on workload management rather than targeting a specific employee segment.

## Q5:How does remote vs office work affect employee productivity and satisfaction, and which setup works better for different roles?

```python
df.head(2)
```

```python
#Compare Productivity (Remote vs Office)
plt.figure(figsize=(6,4))

sns.barplot(x='Remote_Work', y='Productivity_Change_%', data=df)

plt.title('Productivity: Remote vs Office')
plt.xlabel('Work Type (Remote)')
plt.ylabel('Average Productivity Change')

plt.xticks(rotation=0)

# Add labels
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height():.2f}',
                       (p.get_x() + p.get_width()/2, p.get_height()),
                       ha='center', va='bottom')

plt.show()
```

```python
#Compare Satsfaction (Remote vs office)
sns.barplot(x='Remote_Work', y='Job_Satisfaction', data=df)

plt.title('Satisfaction: Remote vs Office')
plt.xlabel('Work Type (Remote)')
plt.ylabel('Average Job Satisfaction')

plt.xticks(rotation=0)

# Add labels
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height():.2f}',
                       (p.get_x() + p.get_width()/2, p.get_height()),
                       ha='center', va='bottom')

plt.show()
```

```python
#Combined View
grouped = df.groupby('Remote_Work')[['Productivity_Change_%', 'Job_Satisfaction']].mean()
print(grouped)
```

```python
#Comparison by job role 
plt.figure(figsize=(10,6))

sns.barplot(x='Job_Role', y='Productivity_Change_%', hue='Remote_Work', data=df)

plt.title('Productivity by Role and Work Type')
plt.xlabel('Job Role')
plt.ylabel('Productivity Change')

plt.xticks(rotation=90)

plt.legend(title='Remote Work')

plt.show()
```

### Conclusion:

Office-based employees show slightly higher productivity (10.17%) compared to remote employees
(9.38%), while job satisfaction remains almost the same in both setups (around 6.0). This indicates that
work location has a minor impact on productivity and no significant impact on satisfaction.

At the role level, there is no consistent pattern where remote or office work is always better. Some roles
perform slightly better in office settings, while others show similar performance in both. This suggests
that the effectiveness of remote work depends on the nature of the job rather than a single general rule.

Overall, work location alone does not strongly determine employee performance. Factors like work
efficiency, tools, and role requirements play a more important role than whether employees work
remotely or in the office.


##### One line answer
Work setup has a limited impact on performance; the role and work efficiency matter more

## Q6:Which departments or roles are at higher risk of job changes, and how can we prepare those employees for future responsibilities?

```python
#For this we will use automation risk column

#Check the percentage that how each job role have a risk to be automated
risk_percent = pd.crosstab(df['Job_Role'], df['Automation_Risk'], normalize='index') * 100 

print(risk_percent)
```

```python
#Plot high risk job roles
high_risk = risk_percent['High'].sort_values(ascending=False)

plt.figure(figsize=(10,5))

high_risk.plot(kind='bar')

plt.title('High Automation Risk by Job Role (%)')
plt.xlabel('Job Role')
plt.ylabel('High Risk Percentage')

plt.xticks(rotation=90)

# Add labels
for i, v in enumerate(high_risk):
    plt.text(i, v, f'{v:.0f}%', ha='center', va='bottom')

plt.show()
```

### Conclusion:
* Risk is higher in roles with repetitive and predictable tasks, while roles requiring human judgment,
creativity, and decision-making show relatively lower risk.

* The analysis shows that roles such as Teacher, SEO Specialist, Store Manager, and Accountant have the
highest percentage of employees at high risk, indicating a greater likelihood of job changes. These roles
often involve structured and repetitive tasks. Roles with medium risk present a strong opportunity for
transition, as employees in these positions can adapt with the right training. To prepare employees for
future responsibilities, organizations should focus on skill development, encourage the use of modern
tools, and shift employees toward more analytical and decision-based work. This will help reduce risk and
improve long-term performance.
